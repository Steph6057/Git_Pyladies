# 8. Sum of numbers divisible by 3 and 5.
# Taking numbers smaller than 10 divisible(without remainder) by 3 or 5, we will get 3, 5, 6, 9. Their sum is 23.What is the sum of the numbers divisible(without remainder) by 3 or 5 in intervals from 0 to 100?

```python
sum = 0

for i in range(0, 100):
  if i % 3 == 0 or i % 5 == 0:
    sum += i

print(sum)
```
